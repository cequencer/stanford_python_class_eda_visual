## Visual inspection
Visual exploration of data is both fun and extremely powerful. It helps you gauge how to best look at the data. 

###Simple graphics is critical
Don't use 3D as it's overkill for 99% of the cases. A simple graphic is like a simple sentence. The message is concise and to the point. A complicated graphic is analogous to a run on sentence. The message of the writer is lost amongst the complexity. 

###Quick and dirty
EDA graphics are meant to be quick and dirty. They give broad oversight on what's happening in the data to explore basic questions. Once understood, they can provide ideas on how to model or apply more complex analysis. Don't go overboard by using D3.js to create a highly interactive plot in the beginning. That step is meant for a final product. 

```python
sns.set_style("white")
fig = plt.figure()
sns.despine(left=True)

dfa = df.dropna()
dfa = df[df.describe().columns]

color = sns.color_palette("husl", dfa.shape[1])


fig = plt.figure(figsize=(10, 16))
for i, col in enumerate(dfa.columns):
    try:
        ax = fig.add_subplot(10, 3, i + 1)
        tmp = dfa[col]
        ax.hist(np.array(tmp), facecolor=color[i], edgecolor='none', histtype='stepfilled')
        ax.set_title(col)
        plt.setp(ax, yticks=[])
        plt.setp(ax, xticks=[])
        sns.despine(left=True)
    except:
        ax.set_title(col)
        plt.setp(ax, yticks=[])
        plt.setp(ax, xticks=[])
        sns.despine(left=True)

plt.tight_layout()
```

**Question:** Is there a correlation between population density and federal spending in throughout US counties? At first glance from the scatter plot below, it seems like the answer is no? Perhaps there is a power-law relationship?

```python
sns.set_style('darkgrid')
plt.scatter(df.density, df.fed_spending, edgecolor='none', facecolor='lightgreen', s=75)
```

Yep, there seems to be a correlation in log-log space. 

```python
sns.set_style('darkgrid')
plt.scatter(np.log10(df.density), np.log10(df.fed_spending), edgecolor='none', facecolor='lightgreen', s=75)
```

**Question:** Is there a relationship between population density and median household income per county? Not likely from the data we have. 

```python
sns.set_style('darkgrid')
plt.scatter(df.density, df.median_household_income, edgecolor='none', facecolor='lightgreen', s=75)
```

In log-log space it does look there is a potential correlation, but we may be seeing an example of projection issues. There may be a variable that allows use disentangle the two primary clumps that we see in the data below. 

```python
sns.set_style('darkgrid')
plt.scatter(np.log10(df.density), np.log10(df.median_household_income), edgecolor='none', facecolor='lightgreen', s=75)
```

```python
#Advanced - but helps see the density of the datapoints
sns.set_style("white")
sns.set_context('talk')

g = sns.JointGrid(np.log10(dfa.density), np.log10(dfa.median_household_income), space=0)
g.plot_marginals(sns.kdeplot, shade=True)
g.plot_joint(plt.scatter, edgecolor='none', facecolor='lightblue', s=40);
```

## Cleaning up data 

Let's assume that you now know what you want from the data. Hence, it's time to clean up the data for the next stage of your analysis.

```python
df_clean = df[['density', 'fed_spending', 'bachelors']]
df_clean = df_clean.dropna() # Remove null values
```

```python
df_clean.describe()
```

```python
sns.set_style("darkgrid")

fig, ax = plt.subplots(figsize=(6, 6))
cmap = sns.blend_palette(["#FF2F08", "#FFFFFF", "#00CAFF"], as_cmap=True)
sns.corrplot(df_clean, annot=False, sig_stars=False,
             diag_names=False, cmap=cmap, ax=ax)

fig.tight_layout()
```
